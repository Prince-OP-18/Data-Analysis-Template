```python
from thefuzz import fuzz,process    
#Similarity for Mutiple String
s1="Hello world"
s2="hello world"
fuzz.ratio(s1,s2)
```

```python
#Extracting similarity from a single list
l1=["Hello World","Programming Great","Coding is necessary","Python","Programming Language","Native programme"]
process.extract("programming",l1,limit=2,scorer=fuzz.partial_token_set_ratio)
```

```python
from rapidfuzz.distance import Levenshtein
s1='cat'
s2='date'
Levenshtein.distance(s1,s2)
```
